# Feature Engineering

```python
counted_stats = [
    'kicks', 'marks', 'handballs', 'disposals',
    'effective_disposals', 'goals', 'afl_fantasy_score', 'SC',
    'behinds', 'hitouts', 'tackles', 'rebounds', 'inside_fifties',
    'clearances', 'clangers', 'free_kicks_for', 'free_kicks_against',
    'contested_possessions', 'uncontested_possessions',
    'contested_marks', 'marks_inside_fifty', 'one_percenters', 'bounces',
    'goal_assists',
    'centre_clearances', 'stoppage_clearances', 'score_involvements',
    'metres_gained', 'turnovers', 'intercepts', 'tackles_inside_fifty',
    'contest_def_losses', 'contest_def_one_on_ones',
    'contest_off_one_on_ones', 'contest_off_wins', 'def_half_pressure_acts',
    'effective_kicks', 'f50_ground_ball_gets', 'ground_ball_gets',
    'hitouts_to_advantage', 'hitout_win_percentage', 'intercept_marks',
    'marks_on_lead', 'score_launches', 'shots_at_goal', 'spoils'
]
```

```python
counted_stat_totals = df.groupby('match_id')[counted_stats].sum().add_suffix('_total')
```

```python
counted_stat_totals
```

```python
s = counted_stat_totals.eq(0).any()
drop_stat_count = list(s[s].index)
```

```python
drop_stat_count
```

```python
df = pd.merge(df, counted_stat_totals, on='match_id', how='inner')
```

```python
df
```

```python
for stat in counted_stats:
    # do not want to divide by 0
    if stat+'_total' not in drop_stat_count:
        df[stat+'_proportion'] = df[stat] / df[stat+'_total']
```

```python
prev_brownlow = pd.read_parquet('../../data/curated/brownlow_votes_22.parquet')
```

```python
# length matches that of dropping duplicates by player_id
# therefore no 2 players have the same full name

# cannot merge via player_id since ids for fryzigg source is different from AFL website
# new players in 2023 do not have assigned fryzigg player_ids

len(prev_brownlow.drop_duplicates('player_id')) == len(prev_brownlow.drop_duplicates(['player_first_name', 'player_last_name']))
```

```python
# Cripps: 29 votes across 21 games => 1.38 average
# looks good
prev = prev_brownlow.groupby(['player_first_name', 'player_last_name'])['brownlow_votes'].mean().sort_values(ascending=False)
prev.head()
```

```python
df = pd.merge(df, prev, 
                  on=['player_first_name', 'player_last_name'],
                  how='left')
```

```python
# all game stats played by new players will have na
df.isna().sum().sort_values(ascending=False)[:4]
```

```python
df.fillna(0, inplace=True)
```

```python
df = df.rename(columns={'brownlow_votes': 'prev_average_votes'})
```
