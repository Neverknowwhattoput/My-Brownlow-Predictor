```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/landing/is_player_captain_23.csv', index_col=0)
```

```python
df
```

```python
df['captain_list'] = df['Captain(s)'].str.split(', ').apply(lambda x: x if isinstance(x, list) else [x])
```

```python
expanded = df['captain_list'].apply(lambda x: pd.Series(x)).fillna('na')

expanded.columns = [f'captain_{i+1}' for i in range(expanded.shape[1])]
```

```python
df = df.join(expanded)[['Club', 'season', 'captain_1', 'captain_2', 'captain_3']]
```

```python
df
```

```python
melted_df = df.melt(
    id_vars=['Club', 'season'],
    value_vars=['captain_1', 'captain_2', 'captain_3'],
    value_name='name'
).drop(columns=['variable'])
```

```python
melted_df = melted_df.query('name != "na"')
```

```python
melted_df
```

```python
players = pd.read_parquet('../../data/raw/player_information_23.parquet')
```

```python
players['name'] = players['player_first_name'] + ' ' + players['player_last_name']
```

```python
final_df = pd.merge(melted_df, players,
         on='name',
         how='left')
```

```python
final_df
```

```python
final_df['is_captain'] = 1
```

```python
final_df.drop(columns=['name', 'Club', 'season'], inplace=True)
```

```python
final_df
```

```python
final_df.to_parquet('../../data/curated/captains_23.parquet')
```
