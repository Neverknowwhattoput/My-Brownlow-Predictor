```python
df = df[cols_keep]
```

```python
matches = pd.DataFrame(df.groupby(['round_number', 'home_team_name']).sum()).sort_values(['round_number', 'home_team_name'])
match_id = [i for i in range(10_000, 10_000 + 23*9)]
matches['match_id'] = match_id
```

```python
matches = matches['match_id']
```

```python
df = pd.merge(df, matches, on=['round_number', 'home_team_name'])
```
