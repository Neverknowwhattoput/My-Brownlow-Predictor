# Investigating 2020-2021 values

```python
df_2020 = df.query('season == 2020')
df_2020_na = df_2020[df_2020.isna().any(axis=1)]
```

```python
df_2020.isna().sum().sort_values(ascending=False)[:2]
```

```python
# afl fantasy scores were not recorded in dataset in all 9 round 1 matches
len(df_2020_na.match_id.unique())
```

```python
idx_to_change = df_2020_na.index
```

```python
# calculate fantasy score via formula on afl website
# https://www.fantasy.afl.com.au/classic/help/game-guidelines

df.loc[idx_to_change, 'afl_fantasy_score'] = 3*df['kicks'] + 2*df['handballs'] \
                                            + 3*df['marks'] +4*df['tackles'] \
                                            + 1*df['free_kicks_for'] \
                                            - 3*df['free_kicks_against'] \
                                            + df['hitouts'] \
                                            + 6*df['goals'] \
                                            + 1*df['behinds']
```

```python
# visible values agree with AFL website
df.loc[idx_to_change][['player_first_name', 'player_last_name', 'afl_fantasy_score']]
```

# Editing 2021 values

```python
df_2021 = df.query('season == 2021')
```

```python
df_2021_na = df_2021[df_2021.isna().any(axis=1)]
```

```python
# only 10 player's rating points are missing
df_2021.isna().sum().sort_values(ascending=False)[:2]
```

```python
col = [
    'match_home_team', 'match_away_team', 'match_round',
    'player_first_name', 'player_last_name', 'rating_points', 'disposals'
]
```

```python
# create table to easily find missing values on AFL website
df_2021_na[col]
```

```python
# fortunately, only 10 errors means it is easy to source by hand
rating_points_in_order = [
    7.3,
    0.5,
    2.8,
    19.8,
    3.3,
    10.6,
    4.8,
    8.5,
    14.3,
    5.5
]
```

```python
idx_to_change = df_2021_na.index
```

```python
df.loc[idx_to_change, 'rating_points'] = rating_points_in_order
```

```python
# check values entered as intended
df.loc[idx_to_change, ['player_last_name', 'rating_points']]
```
